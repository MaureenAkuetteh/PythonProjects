```python
import pandas as pd
import os 
import numpy as np
```

```python
# Read excel file and specific sheet name
filename = 'QC Checks Batch 4 - Nov.xlsx'
sheetname = 'Batch 4'
media = './Media'
outlet_col = 'Outlet id'

excelfile = pd.read_excel(filename, sheet_name = sheetname)
```

```python
#print the first 5 rows of the file
excelfile.head(5)
```

```python
os.listdir?
```

```python
#Get a list of images
images=os.listdir(media)
print(images)
```

```python
#list comprehension
# img_list=[int(image.split('_')[0]) for image in images]

# create a new list to store the image ids 
img_list_2 = []
for image in images :
    img_split = int(image.split('_')[0])
    img_list_2.append(img_split)

    
# print(img_list) 
print(img_list_2)
```

```python
#selecting the outlet id column from the excel file 
excelfile[outlet_col]
```

```python
# drop the rows where the value is NA in the outlet column only 
subset = excelfile.dropna(subset=[outlet_col], inplace=False)
```

```python
#checking if the outlet_id is in the image list 
np.isin(subset[outlet_col],img_list_2)
print(subset)
```

```python
#storing the boolean list in a new column called Has image


subset['Has Image'] = np.isin(subset[outlet_col],img_list_2)
```

```python
subset.head()
```

```python
#ending output to an excel file and removing the count 
#subset.to_csv('image_verification.csv')
subset.to_excel( 'Output/'+ filename, sheet_name= sheetname,index = False)
```
